# Step 3. Perform Item-Based Collaborative Filtering on One Movie Sample

```python
movie_rating_df.head()
```

```python
movie_matrix = movie_rating_df.pivot_table(index = 'user_id', columns = 'movie_title', values = 'rating')
movie_matrix
```

```python
titanic = movie_matrix['Titanic (1997)']
titanic_correlations = pd.DataFrame(movie_matrix.corrwith(titanic), columns = ['Correlation']).reset_index()
```

```python
titanic_correlations = titanic_correlations.join(rating_mean_count_df['count'])
```

```python
titanic_correlations.sort_values('Correlation',ascending=False)
```

```python
titanic_correlations[titanic_correlations['count'] > 80].sort_values('Correlation', ascending=False).head()
```

```python
movie_correlations = movie_matrix.corr(method = 'pearson', min_periods = 80)
movie_correlations
```

```python
myRatings = pd.read_csv('My_Ratings.csv')
```

```python
similar_movies_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlations[myRatings['Movie_Name'][i]].dropna()
    similar_movie = similar_movie.map(lambda x: x*myRatings['Ratings'][i])
    similar_movies_list = similar_movies_list.append(similar_movie)
```

```python
similar_movies_list.sort_values(inplace = True, ascending = False)
print(similar_movies_list.head(10))
```
