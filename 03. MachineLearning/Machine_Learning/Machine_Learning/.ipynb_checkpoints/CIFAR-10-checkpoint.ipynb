# Step 3. Data Preparation

```python
import keras
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
number_cat = 10
```

```python
y_train
```

```python
y_train = keras.utils.to_categorical(y_train, number_cat)  # y_train을 2진법으로 변환함
y_train
```

```python
y_test
```

```python
y_test = keras.utils.to_categorical(y_test, number_cat)  # y_train을 2진법으로 변환함
y_test
```

```python
# 정규화 (dtype = float32로 바꾼 이유)
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
Input_shape = X_train.shape[1:]
Input_shape
```

# Step 4. Train the Model

#### Keras.layers

- Conv2D : 다운 샘플링이나 서브 샘플링을 할 것이라는 의미
- Maxpooling2D : 이미지의 차원 수를 줄이고 실제 특징을 압축하는 방법
- AveragePooling2D : 필터 내 픽셀값의 평균값을 구한다
- Dense : 완전 연결 인공 신경망을 만듦.
- Flatten : 배열을 평평하게 하기 위함
- Dropout : 가중치와 연결된 뉴런을 제거해 정규화함.

```python
# 좌에서 우로 모델을 순차적으로 설계한다는 의미.
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
# 레이어들을 하나씩 하나씩 쌓는 느낌으로 더함.
cnn_model.add(Conv2D(filters=32, 
                     kernel_size=(3,3), 
                     activation='relu',
                     input_shape=(Input_shape)))

# 첫 번째 이후의 레이어들은 첫번째 레이어의 input_shape을 상속받기 때문에 작성할 필요가 없다.
cnn_model.add(Conv2D(filters=32, 
                     kernel_size=(3,3), 
                     activation='relu'))
cnn_model.add(AveragePooling2D(2,2)) # 사이즈를 2x2로  변환함
cnn_model.add(Dropout(0.3)) # 30개의 뉴런을 제거함.

cnn_model.add(Conv2D(filters=64,
                    kernel_size=(3,3),
                    activation='relu'))

cnn_model.add(Conv2D(filters=64,
                    kernel_size=(3,3),
                    activation='relu'))

cnn_model.add(AveragePooling2D(2,2))
cnn_model.add(Dropout(0.2))

cnn_model.add(Flatten())

# 은닉층 생성
cnn_model.add(Dense(units=512, activation='relu')) # 몇 개의 뉴런이 필요한지 알려줘야 한다.
cnn_model.add(Dense(units=512, activation='relu'))

# 출력층 생성
cnn_model.add(Dense(units = 10, activation='softmax')) # 0 or 1, RelU는 연속적인 값을 내뱉기 때문에 사용할 수 없음.
```

### 모델을 컴파일해서 훈련 데이터에 Fit해야함 

```python
cnn_model.compile(loss='categorical_crossentropy', 
                  optimizer = keras.optimizers.RMSprop(lr=0.001),
                  metrics = ['accuracy'])
```

```python
# Batch_size : 신경망에 동시에 몇 개의 이미지를 입력할 것인가?
# shuffle=True : 신경망은 다양한 순서의 이미지에 노출되어야 한다.
history = cnn_model.fit(X_train, y_train, batch_size=32, epochs=2, shuffle=True)
```

# Step 5. Evaluate the Model

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy:{}'.format(evaluation[1]))
```

```python
predicted_classes = cnn_model.predict(X_test)
predicted = predicted_classes.argmax(axis=1)
predicted
```

```python
y_test = y_test.argmax(1)
y_test
```

# Step 7.1. Data Augmentation

```python
import keras
from keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train.shape
```

```python
n = 15
X_train_sample = X_train[:n]
```

```python
X_train_sample.shape
```

```python
# Augmentation
from keras.preprocessing.image import ImageDataGenerator

# 내 이미지를 돌리고 해서 어떻게 재 사용할 수 있도록 할 것인가
dataget_train = ImageDataGenerator(rotation_range = 180,
                                   brightness_range=(10,-1))
dataget_train.fit(X_train_sample)
```

```python
from PIL import Image

fig = plt.figure(figsize=(20,2))

for x_batch in dataget_train.flow(X_train_sample, batch_size=n):
    for i in range(n):
        ax=fig.add_subplot(1,n,i+1)
        ax.imshow(Image.fromarray(x_batch[i].astype(np.uint8))) # 배열을 이미지로 변환
    fig.suptitle('Augmented Images(90 Rotated)')
    plt.show()
    break;
```

# Step 7.2 Model Training Using Augmented Dataset

```python
datagen = ImageDataGenerator(rotation_range = 90,
                             width_shift_range = 0.1,
                             horizontal_flip = True,
                             vertical_flip = True)
```

```python
datagen.fit(X_train)
```

```python
cnn_model.fit_generator(datagen.flow(X_train, y_train, batch_size=32), epochs=2)
```

```python
directory = os.path.join(os.getcwd(), 'saved_models')

if not os.path.isdir(directory):
    os.makedirs(directory)
model_path = os.path.join(directory,'Keras_cifar10_trained_model_Augmentation.h5')
cnn_model.save(model_path)
```
